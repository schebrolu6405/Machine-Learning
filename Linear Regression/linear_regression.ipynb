```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("C:/Users/sushe/Documents/BDA_Fall_2023/BDA_Fall_2023/Sem 2/BDA 602/BDA602_venv/BDA602/House Prices.csv")
data
```

```python
%matplotlib inline
plt.scatter(data.area, data.price, color = 'red', marker = '+')
plt.xlabel('Area(sqr ft)')
plt.ylabel('Price(US$)')
```

```python
#Create Linear regression object
from sklearn import linear_model
reg = linear_model.LinearRegression()
reg.fit(data[['area']], data.price)

```

```python
reg.predict([[3300]])
```

```python
m = reg.coef_
```

```python
b = reg.intercept_
```

```python
y = m*3300 + b
```

```python
y
```

```python
%matplotlib inline
plt.scatter(data.area, data.price, color = 'red', marker = '+')
plt.xlabel('Area(sqr ft)')
plt.ylabel('Price(US$)')
plt.plot(data.area, reg.predict(data[['area']]), color = 'blue')
```

```python
new_df = pd.read_csv("C:/Users/sushe/Documents/BDA_Fall_2023/BDA_Fall_2023/Sem 2/BDA 602/BDA602_venv/BDA602/areas.csv")
new_df.head(3)
```

```python
p = reg.predict(new_df)
new_df['prices'] = p
```
